## 1. Working with Numbers (Integers & Floating-Point Numbers)

**Task:**

1. Create two variables: `x` (an integer) and `y` (a floating-point number).
2. Print their values and types using the `type()` function.
3. Perform addition, subtraction, multiplication, and division between `x` and `y`, then print the results.





```python
x = 12
y = 12

print(type(x))
print(type(y))

print(x + y)
print(x - y)
print(x * y)
print(x / y)
```

## 5. Operators & Expressions

**Task:**

1. Create two variables, `num1 = 15` and `num2 = 4`.
2. Perform the following operations and print the results:
   - Addition (`+`)
   - Subtraction (`-`)
   - Multiplication (`*`)
   - Division (`/`)
   - Floor Division (`//`)
   - Modulus (`%`)
   - Exponentiation (`**`)



```python
num1 = 15
num2 = 4

print(num1 + num2)
print(num1 - num2)
print(num1 * num2)
print(num1 / num2)
print(num1 // num2)
print(num1 % num2)
print(num1 ** num2)
```


---

## 6. Comparison Operators

**Task:**

1. Create two variables, `p = 20` and `q = 10`.
2. Compare `p` and `q` using `==`, `!=`, `>`, `<`, `>=`, and `<=`, then print the results.




```python
p = 20
q = 10

print(p == q)
print(p != q)
print(p > q)
print(p < q)
print(p >= q)
print(p <= q)
```

## 7. String Manipulation

**Task:**

1. Given the string `"Python is awesome!"`, perform the following:
   - Extract the word `"Python"` using slicing.
   - Reverse the entire string.
   - Replace `"awesome"` with `"fun"`.
   - Split the string into words.



```python
string = "Phyton is awesome"
print(string[:6])
print(string[::-1])
print(string.replace('awesome','fun'))
print(string.split())
```
